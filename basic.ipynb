```python
import numpy as np

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])


# 注意这里的操作是元素之间相乘，并不是矩阵相乘。
print(a * b)
print(np.multiply(a, b))

# 使用dot来进行矩阵相乘
print(np.dot(a, b))
```

```python
x = np.array([[1, 2], [3, 4]])

print(np.sum(x))
print(np.sum(x, axis=0))
print(np.sum(x, axis=1))

# convert ndarray to matrix
print(x.T)
```

```python
x = np.array([1, 2, 3])
print(np.reshape(x, (3, 1)))
```

```python
# matplotlib 绘图库
%matplotlib inline # 加入此配置用于显示图像

import matplotlib.pyplot as plt


x = np.arange(-10,10, 0.1)

y = np.cos(x)

plt.plot(x, y)
plt.show()
```

```python
# 下面这两句用来解决中文乱码问题
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x asix label')
plt.ylabel('y asix label')
plt.title(u'正余弦函数图像')
plt.legend(['Sine', 'Cosine'])
plt.show()
```

```python
# 子图

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# 设置子网格高2宽1，并激活绘制 
plt.subplot(2, 1, 1)

plt.plot(x, y_sin)
plt.title('Sine')

plt.subplot(2, 1, 2)

plt.plot(x, y_cos)
plt.title('Cosine')

plt.show()
```
